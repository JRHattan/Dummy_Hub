  # Summary of Data Exploration

Raw dataset of Lending Club loans, spanning from  2007 to 2018 ("accepted_2007_to_2018Q4.csv") pulled from Kaggle.com
Dataset contains over 10 years of loan data, 2.26 million rows of data and 151 columns
There are 63 columns that have over 100,000 missing data points & 94 columns missing over 50,000 values

Loans that have been "Fully Paid" or "Charged Off" is an adequate size for ML models
 * Fully Paid loans: 1,076,751
 * Charged Off loans: 268,559
There appears to be an imbalance between the number of 'good' or "Fully Paid" loans and the 'bad' or "Charged Off" loans

Most data is in "integer" data type, there are 38 columns that are "string" or "date" data types, 
however many of these "string"/"date" columns are also missing large amounts of data ("na")

Much of the missing data will be 'dropped' using the dropna method, and columns/features that 
do not have any prediction power may be dropped after consulting the dataset's data dictionary 
to verify what data they contain and evaluate if the information can be known at the time of loan
application, as well as if the data has no predictive value

Running a correlation table between all of our data columns, we find that there is very high correlation 
between several columns/features which indicates the possibility of duplicate or redundant data
