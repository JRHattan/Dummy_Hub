```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read in CSV data
filtered_data = pd.read_csv("filtered_practdata_LoanStatus_GB.csv", low_memory=False, index_col=False)
filtered_data
```

```python
# Dropping 2 un-needed columns
filtered_data = filtered_data.drop(['Unnamed: 14', 'issue_date'], axis=1)
filtered_data
```

```python
filtered_data.describe()
```

```python
filtered_data["fico_range_low"].describe()
```

```python
#Dropping "na" values
filtered_data = filtered_data.dropna()
```

```python
#Turning column for y/prediction of Good vs. Bad loans into numerical.  1== Good loan. 0 == Bad loan.
filtered_data.good_bad.replace(['Good', 'Bad'], [1, 0], inplace=True)
filtered_data
```

```python
#Setting y/prediction variable column
y = filtered_data['good_bad']
y = y.reset_index()
y = y.drop(columns='index')
y
```

```python
#Setting x/dependent variables to FICO score and dti/Debt-to-Income Ratio columns
X = filtered_data[['fico_range_low', 'dti']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# split Data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
#Instantiate Logistic Regression Model   - KEEP random_state as 1 -
classifier = LogisticRegression(solver='lbfgs', random_state=1)
classifier
```

```python
LogisticRegression(C=1.0, class_weight=None, dual=False, fit_intercept=True,
   intercept_scaling=1, l1_ratio=None, max_iter=100,
   multi_class='warn', n_jobs=None, penalty='12',
   random_state=1, solver='lbfgs', tol=0.0001, verbose=0,
   warm_start=False)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
print(X_train.shape, y_train.shape)
```

```python
#Training the Logistic Regression model
classifier.fit(X_train, y_train)
```

```python
# Validate the Logistic Regression Model
predictions = classifier.predict(X_test)
#pd.DataFrame({"Prediction": predictions, "Actual": y_test})
```

```python
#Accuracy Score
accuracy_score(y_test, predictions)
```
